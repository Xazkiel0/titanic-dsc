```python
import pandas as pd
```

```python
df_train = pd.read_csv('titanic/train.csv')
df_test = pd.read_csv('titanic/test.csv')
```

```python
df_full = pd.concat([df_train, df_test], axis=0, ignore_index=True)
```

```python
sexs = df_train.Sex.unique()
embarkeds = df_train.Embarked.unique()
def sex_norm(sex):
    return sexs.tolist().index(sex)

def embarked_norm(embarked):
    return embarkeds.tolist().index(embarked)

df_train.dropna(subset=['Embarked'],inplace=True)
df_test.dropna(subset=['Embarked'],inplace=True)

df_train.Sex = df_train.Sex.apply(sex_norm)
df_test.Sex = df_test.Sex.apply(sex_norm)
df_train.Embarked = df_train.Embarked.apply(embarked_norm)
df_test.Embarked = df_test.Embarked.apply(embarked_norm)
# harusnya data full train dan test untuk filler mean
df_test.Fare.fillna(df_train.Fare.mean(),inplace=True)
df_train.describe()
```

```python
import math

features = df_train.describe().columns.drop(['Survived']).tolist()

df_train = df_train.loc[:,features+['Survived']]
df_train.Age.fillna(math.ceil(df_train.Age.mean()), inplace=True)
df_test = df_test.loc[:,features]
df_test.Age.fillna(math.ceil(df_test.Age.mean()), inplace=True)
df_test.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(df_train[features], df_train['Survived'], test_size=0.2)

SVM_KERNEL = 'linear'
SVM_C = 10
SVM_GAMMA = 0.00001

svm_model = SVC(kernel=SVM_KERNEL, C=SVM_C, gamma=SVM_GAMMA)
svm_model.fit(X_train, y_train)
```

```python
from sklearn import metrics

y_pred = svm_model.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
